```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib_venn as venn
%matplotlib inline


import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.tools as tls
from plotly.subplots import make_subplots

import plotly.express as px
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode, iplot
from IPython.display import HTML, Image
```

```python
Aquifer_Auser = pd. read_csv("../input/acea-water-prediction/Aquifer_Auser.csv")
Aquifer_Doganella = pd. read_csv("../input/acea-water-prediction/Aquifer_Doganella.csv")
Aquifer_Luco = pd. read_csv("../input/acea-water-prediction/Aquifer_Luco.csv")
Aquifer_Petrignano = pd. read_csv("../input/acea-water-prediction/Aquifer_Petrignano.csv")
```

```python
Aquifer_Auser.columns
```

<a id="1"></a>
# Rainfall

```python
# Aquifer_Auser-Rainfall
Rainfall = Aquifer_Auser.columns
Rainfall = Rainfall[1:11]
data = Aquifer_Auser[Rainfall]
data
```

```python
# Aquifer_Doganella-Rainfall
Rainfall = Aquifer_Doganella.columns
Rainfall = Rainfall[1:3]
data1 = Aquifer_Doganella[Rainfall]
data1
```

```python
# Aquifer_Luco-Rainfall
Rainfall = Aquifer_Luco.columns
Rainfall = Rainfall[1:11]
data2 = Aquifer_Luco[Rainfall]
```

```python
print(Aquifer_Luco[Rainfall])
```

```python
# Aquifer_Petrignano-Rainfall
Rainfall = Aquifer_Petrignano.columns
Rainfall = Rainfall[1:2]
data3= Aquifer_Petrignano[Rainfall]
```

<a id="2"></a>
## Rainfall Concat

```python
rainfall_concat = pd.concat([data,data1,data2,data3],axis =1) 
rainfall_concat.head()
```

```python
rainfall_concat.to_csv("submission.csv", index = False)
```

```python
Rainfall = rainfall_concat.columns
Rainfall = Rainfall[:]
Rainfall
```

```python
traces = [
    go.Histogram(
        x = rainfall_concat[Rainfall[i]].value_counts().unique(), 
        name=Rainfall[i]
    ) for i in range(len(Rainfall))
]

fig = make_subplots(rows=8, cols=3,subplot_titles=())
for i in range(len(traces)):
    fig.append_trace(
        traces[i], 
        (i // 3) + 1, 
        (i % 3) + 1
    )

fig.update_layout(
    template="plotly_white",
    height=1200
)

fig.show()
```
